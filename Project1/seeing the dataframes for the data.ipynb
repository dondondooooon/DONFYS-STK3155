```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('existing.csv')
data
```

```python
plt.figure(figsize=(14, 6), dpi=80)
for i in range(data.shape[0]):
    beta_at_degree = np.array(data.iloc[i])
    beta_at_degree = beta_at_degree[~np.isnan(beta_at_degree)]
    plt.plot(np.arange(0,len(beta_at_degree)),beta_at_degree, label=f'max_deg={i}')
plt.xlabel(r'Polynomial Variable')
plt.ylabel(r'Beta Values $\beta$')
plt.xlim([0,3])
plt.ylim([-100,100])
plt.legend()
plt.show()
```
